```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 2 * np.random.rand(100,1).ravel()
y = 4+3*x+np.random.randn(100,1).ravel()
```

```python
m = b = 0
func = lambda x: m*x + b
```

```python
def plot_line(func, x_points):
    x_points = [i for i in range(int(min(x))-1, int(max(x))+2)]
    y_points = [func(i) for i in x_points]
    plt.plot(x_points, y_points)
```

```python
plot_line(func, x)
```

```python
def summetion(func, x_points, y_points):
    total1 = 0
    total2 = 0
    
    for i in range(1, len(x_points)):
        total1 += func(x_points[i]) - y_points[i]
        total2 += total1 * x_points[i]
    
    return total1/len(x_points), total2/len(x_points)
```

```python
lrn_rate = 0.002
```

```python
plt.scatter(x, y)
for i in range(2000):
    s1, s2 = summetion(func, x, y)
    m = m-lrn_rate*s1
    b = b-lrn_rate*s2
    plot_line(func, x)
plt.show()
```
