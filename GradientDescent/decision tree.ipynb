```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('salaries.csv')
df
```

```python
enc_company = LabelEncoder()
enc_job = LabelEncoder()
enc_degree = LabelEncoder()
```

```python
df.company = enc_company.fit_transform(df.company)
df.job = enc_job.fit_transform(df.job)
df.degree = enc_degree.fit_transform(df.degree)

target = df.salary_more_then_100k

df.drop(['salary_more_then_100k'], axis = 1, inplace = True)
df
```

```python
target
```

```python
model = DecisionTreeClassifier()
model.fit(df, target)
```
